````python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
url1= 'https://www.mathjobs.org/jobs/list/26403'
url2= 'https://www.mathjobs.org/jobs/list/25583'
url3= 'https://www.mathjobs.org/jobs/list/25109'
url_list=[url1,url2,url3]
````

````python
response= rq.get(url1)
response
````

## For One Job

````python
# Missing Values
Name= ''
ID= ''
Title= ''
Type= ''
Location= ''
Area= ''
Deadline= ''
Link= ''

# List Init
Name_List=[]
ID_List=[]
Title_List=[]
Type_List=[]
Location_List=[]
Area_List=[]
Deadline_List=[]
Description_List=[]

soup= bs(response.content)
name= soup.title.text
grid= soup.find('div', class_='grid2 nopd')
divs= grid.find_all('div')
for i in range (0,len(divs),2 ):
  key = divs[i].text
  #print(key)
  if 'ID' in key:
    ID= divs[i+1].text
    #print(ID)
  if 'Title' in key:
    Title= divs[i+1].text
    #print(Title)
  if 'Type' in key:
    Type= divs[i+1].text
    #print(Type)
  if 'Location' in key:
    Location= divs[i+1].text
    #print(Location)
  if 'Area' in key:
    Area= divs[i+1].text
    #print(Area)
  if 'Deadline' in key:
    Deadline= divs[i+1].text
    #print(Deadline)
  if 'Description' in key:
    a_tag= divs[i+1].find('a')
    if a_tag:
      Link= a_tag['href']
      #print(Link)

Name_List.append(name)
ID_List.append(ID)
Title_List.append(Title)
Type_List.append(Type)
Location_List.append(Location)
Area_List.append(Area)
Deadline_List.append(Deadline)
Description_List.append(Link)
````

````python
Data_Dictionary= {'Name':Name_List,
                  'ID':ID_List,
                  'Title':Title_List,
                  'Type':Type_List,
                  'Location':Location_List,
                  'Area':Area_List,
                  'Deadline':Deadline}

df= pd.DataFrame(Data_Dictionary)
df
````

## For All Jobs

````python
url = 'https://www.mathjobs.org/jobs?joblist-0-0------'
response = rq.get(url)
soup = bs(response.content)
div_by_institution_list = soup.find_all('div', class_="clr")
institution_link = []
for div_by_institution in div_by_institution_list:
  li_tags_list = div_by_institution.find_all('li')
  for li_tag in li_tags_list:
    a_tag = li_tag.find('a')
    institution_link.append('https://www.mathjobs.org/'+ a_tag['href'])
list_of_urls =  institution_link
````

````python
# Missing Values
Name= ''
ID= ''
Title= ''
Type= ''
Location= ''
Area= ''
Deadline= ''
Link= ''

# List Init
Name_List=[]
ID_List=[]
Title_List=[]
Type_List=[]
Location_List=[]
Area_List=[]
Deadline_List=[]
Description_List=[]

for url in institution_link:
  response= rq.get(url)
  response

  soup= bs(response.content)
  name= soup.title.text
  grid= soup.find('div', class_='grid2 nopd')

  if grid:
    divs= grid.find_all('div')

    for i in range (0,len(divs),2 ):
      key = divs[i].text
      for i in range (0,len(divs),2 ):
        key = divs[i].text
        #print(key)
        if 'ID' in key:
          ID= divs[i+1].text
          #print(ID)
        if 'Title' in key:
          Title= divs[i+1].text
          #print(Title)
        if 'Type' in key:
          Type= divs[i+1].text
          #print(Type)
        if 'Location' in key:
          Location= divs[i+1].text.split('[')[0]
          #print(Location)
        if 'Area' in key:
          Area= divs[i+1].text
          #print(Area)
        if 'Deadline' in key:
          Deadline= divs[i+1].text
          #print(Deadline)
        if 'Description' in key:
          a_tag= divs[i+1].find('a')
          if a_tag:
            Link= a_tag['href']
            #print(Link)

  Name_List.append(name)
  ID_List.append(ID)
  Title_List.append(Title)
  Type_List.append(Type)
  Location_List.append(Location)
  Area_List.append(Area)
  Deadline_List.append(Deadline)
  Description_List.append(Link)

  Data_Dictionary= {'Name':Name_List,
                  'ID':ID_List,
                  'Title':Title_List,
                  'Type':Type_List,
                  'Location':Location_List,
                  'Area':Area_List,
                  'Deadline':Deadline,
                  'Link': Link}

df= pd.DataFrame(Data_Dictionary)
df
````
